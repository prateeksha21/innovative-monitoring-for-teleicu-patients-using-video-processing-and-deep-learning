```python
# Step 1: Mount Google Drive
from google.colab import drive
drive.mount('/content/gdrive')

# Step 2: Install gdown to download files from Google Drive
!pip install gdown

# Step 3: Download the video from Google Drive
import gdown

# Replace with your actual file ID
file_id = '1-Kt0k7WM6t6ejpfKY3ymTi6aZE4GRdR5'
gdown.download(f'https://drive.google.com/uc?export=download&id={file_id}', 'TEST_ICU.mp4', quiet=False)

# Step 4: Function to extract frames from the video
import cv2
import os

def extract_frames(video_file, output_directory, frame_rate=2):
    cap = cv2.VideoCapture(video_file)
    frame_count = 0
    os.makedirs(output_directory, exist_ok=True)

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        frame_count += 1
        if frame_count % int(cap.get(cv2.CAP_PROP_FPS) / frame_rate) == 0:
            output_file = f"{output_directory}/frame_{frame_count}.jpg"
            cv2.imwrite(output_file, frame)
            print(f"Frame {frame_count} has been extracted and saved as {output_file}")

    cap.release()
    cv2.destroyAllWindows()

# Step 5: Extract frames from the downloaded video and save to the TELEICU_frames folder in Google Drive
video_file = 'TEST_ICU.mp4'
output_directory = '/content/gdrive/MyDrive/Intel unnati project/TELEICU_frames'
extract_frames(video_file, output_directory)
```
